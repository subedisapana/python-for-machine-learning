#### Model Evaluation

We can use confusion matrix to evaluate our model.
Imagine testing for disease whether there is presence of disease or not.

Let's say: You are going to test whether a women is pregnent or not. If yes, the test is positive or else negative.

- __True Positive(TP)__: These are cases in which test predicts positive and in actual also pregnant(right prediction)

- __True Negative(TN)__: These are cases in which test predicts negative and in actual also not pregnant (wrong prediction)

- __False Positive(FP)__: Test showing pregnency test positive for a man. (wrong prediction) __Type I Error__

- __False Negative(FN)__: A pregnant women with bumps tests for -pregnancy and test shows negative. (wrong prediction) __Type II Error__

#### Accuracy 

- Overall, how often is it correct? (TP+TN)/total
- Overall, how often is it incorrect? (FP+FN)/total


#### Precision

total number of correctly classified positive cases / ( total number of positive predictated cases)

__Positive predictive values = TP/(TP+FP)__

total number of correctly specified negative cases / (total number of negative cases)

__Negative predictive value = TN/(TN+FN)__


#### Recall (Sensitivity)

the ratio of the total number of correctly classified positive cases /  the total number of positive cases. High Recall indicates the class is correctly recognised (small number of FN).

__Recall = TP/(TP+FN)__


#### Specificity:  

AKA True Negative Rate. Measures the proportion of Actual Negatives that are correctly identified as such (e.g. the percentage of women cases correctly identified as not having the baby)

__Specificity = TN/(TN+FN)__

#### F1-score

It is difficult to compare two models with low precision and high recall or vice versa. So to make them comparable, we use F1-Score. F1-score helps to measure Recall and Precision at the same time. It uses Harmonic Mean in place of Arithmetic Mean by punishing the extreme values more.

__F1-Score = 2* (Recall * Precision)\ (Recall + Precision)__
